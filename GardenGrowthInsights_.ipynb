# GardenGrowthInsights:

```python
import csv

#define your data 

data = [['Day' , 'FlowerCount' , 'Hight'],
    [1, 5, 10],
    [2, 8, 15],
    [3, 12, 20],
    [4, 18, 25],
    [5, 22, 30]]

#simplfy the filename

file_name = 'garden-data.csv'

# Write to CSV file
with open(file_name , 'w' , newline= '') as csvfile:
    csvwriter = csv.writer(csvfile)
    csvwriter.writerows(data)
    
print(f'CSV file "{file_name}"  is created Succeefully.')
```

# IMPORTING THE DATA SET

```python
import pandas as pd 

# Read the dataset

file_name = 'garden-data.csv'
df = pd.read_csv(file_name)


# Display the initial state of the DataFrame
print("Initial DataFrame:")
print(df)
```

# DATA CLEANING

```python
# Check for missing values
missing_value = df.isnull().sum()
print("\nMissing Values:")
print(missing_value)
```

```python
# Handle missing values (replace or remove)
# For simplicity, let's fill missing values with the mean of the respective column
df.fillna(df.mean() , inplace = True)
```

```python
# Check for negative heights and handle them
# Assuming heights should be non-negative, we'll replace negative heights with the absolute value
df['Hight'] = df['Hight'].apply(lambda x : abs(x) if x < 0 else x)
```

```python
# Display the cleaned DataFrame
print("\nCleaned DataFrame:")
print(df)
```

# visualization 

```python
import pandas as pd
import matplotlib.pyplot as plt 
```

```python
# Read the dataset
file_name = 'garden-data.csv'
df = pd.read_csv(file_name)
```

```python
# Data Visualization

plt.figure(figsize=(10,6))
```

```python
# Line plot for FlowerCount

plt.plot(df['Day'] , df['FlowerCount'], marker = 'o' , label = 'Flower Count' , color = 'Blue')
# Add labels and title
plt.xlabel('Days')
plt.ylabel('Count/Height')
plt.title('Garden Growth Over Days')
plt.legend()
plt.grid()
```

```python
# Line plot for Height

plt.plot(df['Day'] , df['Hight'] , marker = 'o' , label = 'Hight' , color = 'b')

# Add labels and title
plt.xlabel('Days')
plt.ylabel('Count/Height')
plt.title('Garden Growth Over Days')
plt.legend()
plt.grid()
```
